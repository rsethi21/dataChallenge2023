```python
import pandas as pd 
```

```python
female_data_path = "/Users/johnfeng/Documents/Data_challenge/Data/processed_female_data.csv"
male_data_path = "/Users/johnfeng/Documents/Data_challenge/Data/processed_male_data.csv"

female_data = pd.read_csv(female_data_path)
male_data = pd.read_csv(male_data_path)
```

```python
usa_female_data = female_data[female_data['Country'] == 'USA']
usa_male_data = male_data[male_data['Country'] == 'USA']

usa_female_data.to_csv('usa_processed_female_data.csv', index=False)
usa_male_data.to_csv('usa_processed_male_data.csv', index=False)
```

```python
usa_female_names = usa_female_data['names'].unique()
usa_male_names = usa_male_data['names'].unique()

usa_female_names_series = pd.Series(usa_female_names)
usa_male_names_series = pd.Series(usa_male_names)

usa_female_names_series.to_csv('usa_female_namelist.csv', index=False, header=['Name'])
usa_male_names_series.to_csv('usa_male_namelist.csv', index=False, header=['Name'])
```
